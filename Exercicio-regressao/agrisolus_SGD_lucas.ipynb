# Missing Data

```python
def missing_data(df):
    total = df.isnull().sum().sort_values(ascending = False)
    percentual = round(df.isnull().sum().sort_values(ascending = False) / len(df) * 100, 2)
    return pd.concat([total, percentual], axis=1, keys=['Total','Percentual'])
```

```python
#verified nan data
missing_data(df_agrisolus_dias)
```

# <b>RSS</b> and <b>R$^2$<b>

```python
from sklearn.metrics import r2_score
def calc_rss(y,predicted):
    return float(((predicted - y) ** 2).sum())
    
def calc_r2(y,predicted):
    return r2_score(predicted,y)
```

# Linear Regression

```python
from sklearn.model_selection import train_test_split

X = df_agrisolus_dias.loc[:,['Idade']]
y = df_agrisolus_dias.loc[:,['Peso']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, shuffle=True)
```

```python
from sklearn import linear_model

X = X_train.loc[:, ['Idade']]
y = y_train.loc[:, ['Peso']]

reg = linear_model.LinearRegression()

y_hat = reg.fit(X, y).predict(X)
```

```python
#RSS and R2

rss = calc_rss(y,y_hat)
r2 = calc_r2(y,y_hat)
print('RSS:', rss)
print('R2:',r2)
```

# SGD Regression

```python
from sklearn.model_selection import train_test_split

#X = df_agrisolus_dias.loc[:,['Idade', 'Temperatura', 'Umidade']]
#X = df_agrisolus_dias.loc[:,['Idade', 'Temperatura']]
#X = df_agrisolus_dias.loc[:,['Idade', 'Umidade']]
X = df_agrisolus_dias.loc[:,['Idade']]
y = df_agrisolus_dias.loc[:,['Peso']]

X_SGD_train, X_SGD_test, y_SGD_train, y_SGD_test = train_test_split(X, y, test_size=0.15, shuffle=True)
```

```python
from sklearn.linear_model import SGDRegressor

X_SGD = X_SGD_train
y_SGD = y_SGD_train

final = 9
for k in range(1, final): 
    reg_grad = SGDRegressor(max_iter=5000,alpha=0)
    y_SGD_hat = reg_grad.fit(X_SGD, y_SGD.values.ravel())

    y = y_SGD
    y_hat = y_SGD_hat.predict(X_SGD).reshape(-1,1)

    rss = calc_rss(y,y_hat)
    r2 = calc_r2(y,y_hat)
    print('k:', k)
    print('RSS:', rss)
    print('R2:',r2, '\n')
```

# SGD Regression `alpha`

```python
from sklearn.model_selection import train_test_split

#X = df_agrisolus_dias.loc[:,['Idade', 'Temperatura', 'Umidade']]
#X = df_agrisolus_dias.loc[:,['Idade', 'Temperatura']]
#X = df_agrisolus_dias.loc[:,['Idade', 'Umidade']]
X = df_agrisolus_dias.loc[:,['Idade']]
y = df_agrisolus_dias.loc[:,['Peso']]

X_SGD_train, X_SGD_test, y_SGD_train, y_SGD_test = train_test_split(X, y, test_size=0.15, shuffle=True)
```

```python
from sklearn.linear_model import SGDRegressor

X_SGD = X_SGD_train
y_SGD = y_SGD_train

final = 4

for k in range(1, final):
    random_number = np.random.random()
    reg_grad = SGDRegressor(max_iter=5000,alpha=random_number)
    y_SGD_hat = reg_grad.fit(X_SGD, y_SGD.values.ravel())

    y = y_SGD
    y_hat = y_SGD_hat.predict(X_SGD).reshape(-1,1)

    rss = calc_rss(y,y_hat)
    r2 = calc_r2(y,y_hat)
    print('k:', k)
    print('coeficient:', reg_grad.coef_)
    print('alpha:', random_number)
    print('RSS:', rss)
    print('R2:',r2)
    print('MSE:', mean_squared_error(y, y_hat), '\n')
```
